```python
import pandas as pd
```

```python
file_name = 'tourism.xlsx.xlsm'

all_data = pd.read_excel(file_name, sheet_name=None, engine='openpyxl')

print("sheet_list", all_data.keys())
```

```python
df1 = all_data["Sheet1"]

df1.head()
```

```python
rename_map = {
    '通し番号': 'id',
    '都道府県': 'prefecture',      
    '日本語タイトル': 'name',
    "日本語本文": "explanation"  
}

df2 = df1[list(rename_map.keys())].rename(columns=rename_map)
print(df2.columns)
```

```python
df2.head()
```

```python
#prefecture unique numbers  (originally it was  96? 95)
regions = df2["prefecture"].unique()
print(len(regions))
```

```python
print(list(regions))
```

```python
#space delete
df2['prefecture'] = df2['prefecture'].str.strip()

#(\n) delete
df2['prefecture'] = df2['prefecture'].str.replace('\n', '', regex=False)

# 3. space delete
df2['prefecture'] = df2['prefecture'].str.replace(' ', '', regex=False)


print(df2['prefecture'].unique())
```

```python
# add "kenn"
fix_map = {
    '鹿児島': '鹿児島県',
    '和歌山': '和歌山県'
}

df2['prefecture'] = df2['prefecture'].replace(fix_map)


# prefecture names united
def sort_names(text):
    if isinstance(text, str) and "・" in text:
        parts = text.split("・")
        parts.sort()
        return "・".join(parts)
    else:
        return text
    
df2["prefecture"] = df2["prefecture"].apply(sort_names)


# output 
prefecture_list_final = sorted(df2['prefecture'].unique())
print(prefecture_list_final)
print(len(prefecture_list_final))
```

```python
# id numbers
len(df2["id"])
```

```python
# duplicated numbers of tourist attractions number
print(df2["name"].duplicated().sum())
```

```python
df3 = df2.sort_values(by ="explanation", key = lambda x: x.str.len(), ascending= False).drop_duplicates(subset=["name"])

print(len(df3["name"]))
```

```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform

system_name = platform.system()

if system_name == 'Windows':
    rc('font', family='Malgun Gothic') 
elif system_name == 'Darwin': 
    rc('font', family='AppleGothic')
else:
    rc('font', family='NanumGothic')

plt.rcParams['axes.unicode_minus'] = False

df_grouped = df3.groupby('prefecture')['name'].count().reset_index()

df_grouped = df_grouped.sort_values(by='name', ascending=False)

plt.figure(figsize=(12, 6)) 
sns.barplot(data=df_grouped, x='prefecture', y='name')

plt.title('Prefecture Count', fontsize=15)
plt.xlabel('Prefecture', fontsize=12)
plt.ylabel('Count of Names', fontsize=12)
plt.xticks(rotation=90) 

plt.show()
```
